# (very) basic usage of numpy

````python
# import numpy for basic math functions
# access functions via: np.sin() , np.cos(), np.sqrt(), ...
import numpy as np
````

````python
# numpy comes with it's own array-type called 'numpy-array' 
# it can be used to perform element-wise arithmetic operations

# initialize a numpy array
x = np.array([.1, .2, .3])
````

````python
# --- perform some calculations
y = x
print('x = ', y, '\n')
y = x + 5
print('x + 5 = ', y, '\n')
y = x**2
print('x^2 = ' , y, '\n')
y = np.sin(x)
print('sin(x) = ', y, '\n')
y = np.cos(x)
print('cos(x) = ', y, '\n')

# --- numpy array-broadcasting
x2 = np.array([x, 2*x, 3*x])
print('x2 = \n', x2, '\n')
y = x + x2
print('x + x2 = \n', y, '\n')
x3 = np.array([[1],[2],[3]])
y = x + x3
print('x3 = \n', x3, '\n')
print('x + x3 = \n', y, '\n')
````

# basic (static) plot generation with matplotlib

````python
# import numpy for basic math functions
import numpy as np
# import the matplotlib.pyplot plotting library
import matplotlib.pyplot as plt
# set matplotlib-backend to 'inline' to get simple static plots
%matplotlib inline
````

````python
# generate a simple static plot

# generate 20 evenly spaced numbers between 0. and 1.
x = np.linspace(0., 1., 20)
# calculate the square root of x
y = np.sqrt(x)

# --- begin plotting ---

# generate a figure-object with specific size
fig = plt.figure(figsize = (10, 5))
# add an axis
ax = fig.add_subplot(111)

# --- some possibilities for lines and points ---

# plot a solid line
ax.plot(x, y)
# plot a dashed line and give it a label
ax.plot(x, y + 0.5, '--', label = 'this is a dashed line')
# plot a yellow line with increased line-width and 50% opacity
ax.plot(x, y + 1, 'y-', linewidth = 5, alpha = 0.5)
# plot a line with rather big point-markers
ax.plot(x, y + 1.5, '-', marker = '.', markersize = 20, label = 'a line with big markers')
# plot points
ax.plot(x, y + 2, '.')
# plot larger green points
ax.plot(x, y + 2.5, 'go')
# plot triangles
triang,= ax.plot(x, y + 3, '>')
# plot another set of triangles at the same color as the line before
ax.plot(x, y + 3.5, '<', color = triang.get_color())

# --- customization of the plot ---

# set custom boundaries for x- and y- axis
ax.set_xlim([-.2, 1.2])
ax.set_ylim([-1., 6])

# labelling of axes and plots
ax.set_title('this is the figure-title')
ax.set_xlabel('here´s the label of the x-axis')
ax.set_ylabel('and the $\sqrt{y_{axis}}$ uses LaTex syntax!')

# generate a legend for all labelled objects in ax
ax.legend()

# show the plot
plt.show()
````

# Basic (dynamic) plot generation with matplotlib

### interactive Matplotlib backend
To get a interactive figure (instead of a static png) with matpltolib you must switch to an interactive backend!
- this is done by executing `%matplotlib widget`

###  Ipython widgets
- the most simple interaction can be achieved with `interact`
    
    ```python
    from ipywidgets import interact
    ```


- it can be used as a function `interact(f, args)` or a decorator `@interact(args)`

| args | What you get |
| --- | --- |
| `x = "a string"` | a text-input form |
| `x = (start, stop, [step])` | a slider (either 'int' or 'float') |
| `x = ["first item", "second item"]` | a dropdown list |



  - there are a lot more widgets to explore! (see [Widget List](https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20List.html#Widget-List))
     - buttons, sliders, boxes, selectors
     - animation (e.g. Play)
  - control multiple functions with a single widget (see [Linking two similar widgets](https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20Basics.html#Linking-two-similar-widgets))
  - control how your widgets are displayed  (see [Layout and Styling of Jupyter widgets](https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20Styling.html#Layout-and-Styling-of-Jupyter-widgets))


for more details have a look at:
- https://ipywidgets.readthedocs.io/en/latest/examples/Using%20Interact.html
- https://ipywidgets.readthedocs.io/en/latest/examples/Widget%20List.html


````python
# import the ipython interact-widget to get nice sliders etc.
from ipywidgets import interact
# you must have set matplotlib-backend to 'widget' to support interactive plots!
%matplotlib widget
````

````python
# generate a simple dynamic plot
# initialize arrays for plotting
x = np.linspace(0., 15., 50)
y = np.cos(x)

# --- generate your plot as usual, but this time provide reference
#     to all lines etc. that are intended to be dynamically updated

# initialize figure object
fig = plt.figure(figsize=(10, 3))
ax = fig.add_subplot(111)
ax.set_title('This is a bad figure... it has no labels, legends etc..!')

# plot a line that will later be updated and name it l1
# (the comma after l1 is needed since ax.plot() returns a list)
# (e.g. this is the same as   lines = ax.plot(...)    l1 = lines[0] )
l1, = ax.plot(x, y)
# plot a line that remains static
ax.plot(x, y, '--', alpha = 0.5)

# --- setup dynamic plot ---

# define a function that updates the y-data of the line l1
# and make it interactive by using the @interact decorator
@interact(a=(-2.*np.pi, 2.*np.pi))
def update(a):
    l1.set_ydata(np.cos(a + x))
````

````python
# it is also possible to put everything in a function and execute the function in a later celL!
def makeafigure():
    x = np.linspace(0., 15., 50)
    y = np.cos(x)

    fig = plt.figure(figsize=(10, 3))
    ax = fig.add_subplot(111)
    ax.set_title('This is a bad figure... it has no labels, legends etc..!')

    l1, = ax.plot(x, y)
    ax.plot(x, y, '--', alpha = 0.5)

    def update(a):
        l1.set_ydata(np.cos(a + x))

    out = interact(update, a=(-2.*np.pi, 2.*np.pi))
    return out
````

````python
figure1 = makeafigure()
````

# a bit more advanced (dynamic) plot generation with matplotlib
the following example shows a few more possibilities for generating
interactive figures

````python
# (see http://ipywidgets.readthedocs.io/en/stable/examples/Widget%20List.html 
# for more details on how to customize sliders)
from ipywidgets import widgets, IntSlider, FloatSlider, Dropdown
````

````python
# a dynamic plot-widget
def make_another_figure():
    # generate a figure
    fig = plt.figure(figsize=(12, 4))
    fig.subplots_adjust(top=.9, bottom=0.1, hspace=1)

    # generate axis-object for the first plot
    ax = fig.add_subplot(211)
    ax.set_ylim(-10, 10)
    ax.set_xlabel('x')
    ax.set_ylabel('$f(x)$')

    # generate axis-object for the second plot
    ax2 = fig.add_subplot(212)
    ax2.set_ylim(-5, 15)
    ax2.set_xlabel('x')
    ax2.set_ylabel('$f(x)$')


    x = np.linspace(-10, 10, num=100)
    y = np.sin(x)

    # plot lines
    line1, = ax.plot(x, y, 'r-', marker = '.', label = 'line 1') 
    line11, = ax.plot(x, np.zeros_like(y), '--', label = 'line 11') 
    line2, = ax2.plot(x, y**2, label = 'line 2')

    # fill between x and the line of the second plot
    fill = {'fill':ax2.fill_between(x, y**2, alpha = 0.5)}

    ax.legend()
    ax2.legend()

    title1 = (r'here you can see a plot of: \n f(x) = ...')
    ax.set_title(title1)

    title2 = (r'...and here of of : ...')

    ax2.set_title(title2)


    # define a function to update the plot
    def func(x0, A, f, omega, N):
        if   f == "sin": func = np.sin
        elif f == "cos": func = np.cos
        elif f == "tan": func = np.tan

        # define updated plot-arrays
        y = x0 + A*func(omega * x) 
        y2 = y**N

        # update plotted lines
        line1.set_ydata(y)
        line1.set_label(rf'$f(x) = x_0 + A \cdot {f} (\omega \cdot x)$')

        line11.set_ydata(np.full_like(y, x0))
        line11.set_label(rf'$f(x) = x_0 = {x0}$')
        line2.set_ydata(y2)
        line2.set_label(rf'$f(x) = [\, x_0 + A \cdot {f} (x)\, ]^{N}$')

        # update titles and labels
        title1 = f'here you can see a plot of: \n $f(x) = {x0} + {A} * {f}({omega} * x)$' 
        ax.set_title(title1)

        title2 = f'...and here of of : \n $f(x) = [{x0} + {A} * {f}({omega} * x)]^{N}$' 
        ax2.set_title(title2)

        # remove and re-draw the filling in the second plot
        fill['fill'].remove()
        fill['fill'] = ax2.fill_between(x, y2, alpha = 0.5, color='C0')

        # re-generate legends (located at position 1, i.e. the upper right corner)
        ax.legend(loc='upper right')
        ax2.legend(loc='upper right')    


    # setup the slider objects
    x0 = FloatSlider(min=-2, max=3, description='x0', value=1)
    A = FloatSlider(min=0., max=3., description='A', value=1)
    omega = FloatSlider(min=0., max=2., value=1.2, description=r'$\omega$')
    f = Dropdown(options=['sin', 'cos', 'tan'], value='sin', description='func')
    N = IntSlider(min=2, max=5, description='N', value=3)

    # arrange the sliders in a box
    ui = widgets.HBox([widgets.VBox([A, omega, N]), widgets.VBox([x0, f])])

    # generate the widget
    out = widgets.interactive_output(func, {'x0': x0, 'A': A, 'omega': omega, 'f':f, 'N':N})
    return display(out, ui)
````

````python
figure2 = make_another_figure()
````

## control multiple functions with a single widget

````python
# interact with multiple functions
from ipywidgets import widgets, interactive_output

slider1 = widgets.FloatSlider(value=1, min=0, max=10, step=0.5, description="the x slider")
slider2 = widgets.FloatSlider(value=5, min=0, max=10, step=0.5, description="the y slider")
ui = widgets.HBox((slider1, slider2))

def f(x, y):
    print(f"first function: \n x = {x:.2f}   y = {y}")

def f2(x, y):
    print(f"second function: \n x**2 = {x**2:.2f}   y**2 = {y**2}")


out = interactive_output(f, dict(x=slider1, y=slider2))
out2 = interactive_output(f2, dict(x=slider1, y=slider2))

outbox = widgets.HBox((out, out2), 
                     layout=dict(border="solid 2px",
                                 grid_gap="100px",
                                 width="50%"))

display(ui, outbox)
````
