# Motivation: Maxwell's Equations in Matter

Maxwell's equation describe the behavior of electromagnetic fields in matter. In differential form, they are given by:

$\nabla \times \vec{E}\left(\vec{r},t\right) = -\dfrac{\partial }{\partial t}\vec{B}\left(\vec{r},t\right)$

$\nabla \times \vec{H}\left(\vec{r},t\right) = \vec{j}_{ext}\left(\vec{r},t\right) + \dfrac{\partial }{\partial t}\vec{D}\left(\vec{r},t\right)$

$\nabla \cdot \vec{D}\left(\vec{r},t\right) = \rho_{ext}\left(\vec{r},t\right)$

$\nabla \cdot \vec{B}\left(\vec{r},t\right) = 0$


When dealing with matter, Maxwell's equation can be expressed in terms of the four macroscopic fields:

* Electric field $\vec{E}\left(\vec{r},t\right)$
* Magnetic flux density $\vec{B}\left(\vec{r},t\right)$
* Electric displacement field $\vec{D}\left(\vec{r},t\right)$
* Magnetic field $\vec{H}\left(\vec{r},t\right)$

Further, these equations depend on the external charge density $\rho_{ext}\left(\vec{r},t\right)$ and the current density $\vec{j}_{ext}\left(\vec{r},t\right)$ and are linked to the material properties of the medium, namely the polarization $\vec{P}\left(\vec{r},t\right)$ and magnetization $\vec{M}\left(\vec{r},t\right)$, by the following relations:

$\vec{D}\left(\vec{r},t\right) = \epsilon_0\epsilon_r \vec{E}\left(\vec{r},t\right) = \epsilon_0\vec{E}\left(\vec{r},t\right) + \vec{P}\left(\vec{r},t\right)$

$\vec{B}\left(\vec{r},t\right) = \mu_0\mu_r \vec{H}\left(\vec{r},t\right) = \mu_0\vec{H}\left(\vec{r},t\right) + \vec{M}\left(\vec{r},t\right)$


From here on, the time and position dependency of the fields will be omitted for the sake of simplicity. Further, both the relative permittivity $\epsilon_r$ and the relative permeability $\mu_r$ are complex and functions of $\omega$.

Solving Maxwell's equation yields an expression for the wave equation. 
  
For an electromagentic wave (emw) originated at location $\vec{r_0}$, of free-space wavelength $\lambda_0$ and propagating into direction $\vec{k}$ observed in the far field (hence a plane wave), the following expression can be used to describe the electric field associated with this plane emw:

$\vec{E} = \vec{E_0} \cdot \exp\left( i \left( \vec{k}\vec{r} - \omega t \right) \right)$

$\vec{E_0}$ denotes the electric field amplitude, which is chosen to be kept complex, the angular frequency $\omega = 2\pi\nu = 2\pi \cdot \dfrac{\mathrm{c}}{\lambda} = \mathrm{c}\cdot k$ (with the speed of light in vacuum $\mathrm{c}$ and the wavenumber $k$) and $t$ is the time.

Further, the speed of light in vacuum $\mathrm{c}$ is related to the vacuum electric permittivity $\epsilon_0$, the vacuum magnetic permeability $\mu_0$ and the free-space wavelength and frequency $\lambda_0$ and $\nu_0$ by:
$\mathrm{c} = \dfrac{1}{\sqrt{\epsilon_0 \mu_0}} = \lambda_0\nu_0$. 

Inside a medium with $\epsilon_r, \mu_r > 1$, the propagation speed of the emw is the speed of light normalized py the mediaum's properties:

 $v_g = \dfrac{1}{\sqrt{\epsilon_r \epsilon_0 \mu_r \mu_0}} = \dfrac{1}{\sqrt{\epsilon_r \mu_r}} \cdot \dfrac{1}{\sqrt{\epsilon_0 \mu_0}} = \dfrac{\mathrm{c}}{\sqrt{\epsilon_r \mu_r}} = \dfrac{\mathrm{c}}{n_c}$.

The proportionality factor $\sqrt{\epsilon_r \mu_r}$ is called the complex refractive index $n_c$ of the medium. Just like the relative permittivity, it can be be expressed as (if the medium is non-magnetic, i.e. $\mu_r = 1$):

$\epsilon_r(\omega) = \epsilon_r'(\omega) + i \epsilon_r''(\omega) = n_c^2(\omega)$

$n_c(\omega) = n(\omega) + i\kappa(\omega)$

The real part $n(\omega)$ is called the refractive index and the imaginary part $\kappa(\omega)$ is called the extinction coefficient. The refractive index is a measure for the phase velocity of the emw in the medium, while the extinction coefficient is a measure for the attenuation of the emw in the medium.

## 1.1.1 How does the wave-equation change if we consider an electromagnetic wave traveling inside conducting, linear material?

As the lecture already showed, by solving Maxwell's equation inside a medium, the wavenumber indeed is of complex nature. Thus, here a qualitative approach is chosen to derive the same result.

It was shown above, that the propagating speed of an emw depends on the material's properties it is propagating through. According to $v_g = \dfrac{\mathrm{c}}{n_c}$ and $\mathrm{c} = \nu_0 \cdot \lambda_0$, the wavelength also depends on the material's properties. Hence, so does the wavenumber $k$. For a non-magnetic, isotropic, conductive and lossy medium follows:

$v_g = \dfrac{\mathrm{c}}{n_c} = \dfrac{\nu_0 \cdot \lambda_0}{n_c} \Rightarrow \lambda = \dfrac{\lambda_0}{n_c}$

$k = \dfrac{2\pi}{\lambda} = k_0 \cdot n_c = k_0n + ik_0\kappa$

Simplifying the expression for the electric field in such a way, that the wave is propagating into $x$-direction in this medium finally yields:

$\vec{E}(x, t) = \vec{E_0} \cdot \exp\left( i \left( kx - \omega t \right) \right) = \vec{E_0} \cdot \exp\left( i \left(k_0(n + i\kappa)x - \omega t \right) \right) = \vec{E_0} \cdot \exp\left( i \left( k_0nx - \omega t \right) \right) \cdot \exp\left( -k_0\kappa x \right)$

The first term is the same as for a non-conducting medium, while the second term is an exponential decay term, which describes the attenuation of the emw in the medium. The decay constant is given by $k_0\kappa$. 

Briefly summarized:
* By propagating through a conducting medium, the wave is attenuated by a factor of $\exp\left( -k_0\kappa x \right)$ per unit length $x$
* This attenuation is partially the result of the medium's permitivity and conductivity $\sigma$ (vacuum: $\sigma = 0$), which gives rise to a current density according to Ohm's law and is accounted for in the Maxwell equations above: $\vec{j} = \sigma \vec{E}$
* Through the change of dielectric environment (vacuum vs. medium), the propagagtion speed of the wave decreases and the wavelength, too

**For the sake of simplicity, in the following example the wave is assumed to not propagate, but 'frozen in time' at $t = 0$.**

```python
def emw(x, lambda_0: float, n: float, kappa: float, amplitude: Optional[float] = 1) -> float:
    '''Simple EMW function to calculate the electric field strength at a given position x.
    Parameters:
    -----------
    x: float
        Position in m
    lambda_0: float
        Wavelength in m
    n: float
        Refractive index
    kappa: float
        Absorption coefficient
    amplitude: float, optional
        Amplitude of the EMW, default is 1

    Returns:
    --------
    float
    '''
    
    k0 = 2 * np.pi / lambda_0
    return (amplitude * np.exp(1j * k0 * n * x) * np.exp(-k0 * kappa * x)).real
```

```python
# %matplotlib inline
%matplotlib widget

x = np.linspace(0., 15., 50)
y = np.cos(x)

# --- generate your plot as usual, but this time provide reference
#     to all lines etc. that are intended to be dynamically updated

# initialize figure object
fig = plt.figure(figsize=(10, 3))
ax = fig.add_subplot(111)
ax.set_title('This is a bad figure... it has no labels, legends etc..!')

# plot a line that will later be updated and name it l1
# (the comma after l1 is needed since ax.plot() returns a list)
# (e.g. this is the same as   lines = ax.plot(...)    l1 = lines[0] )
l1, = ax.plot(x, y)
# plot a line that remains static
ax.plot(x, y, '--', alpha = 0.5)

# --- setup dynamic plot ---

# define a function that updates the y-data of the line l1
# and make it interactive by using the @interact decorator
@interact(a=(-2.*np.pi, 2.*np.pi))
def update(a):
    l1.set_ydata(np.cos(a + x))
```

```python

x = np.linspace(0., 6*np.pi, 100)
y = emw(x = x, lambda_0 = 0.19, n = 1.5, kappa = 0.01, amplitude = 1)

fig = plt.figure(figsize=(10, 3))
ax1 = fig.add_subplot(121)
ax1.set_title(f'Electric field part of an electromagentic wave propagating in a medium of refractive index n = {1.5} and absorption coefficient kappa = 0.01')

# plot a line that will later be updated and name it l1
# (the comma after l1 is needed since ax.plot() returns a list)
# (e.g. this is the same as   lines = ax.plot(...)    l1 = lines[0] )
l1, = ax1.plot(x, y)
# plot a line that remains static
ax1.plot(x, y, '--', alpha = 0.5)

# --- setup dynamic plot ---

# define a function that updates the y-data of the line l1
# and make it interactive by using the @interact decorator
@interact(a=(-2.*np.pi, 2.*np.pi))
def update(a):
    l1.set_ydata(np.cos(a + x))
# plt.show()
```
